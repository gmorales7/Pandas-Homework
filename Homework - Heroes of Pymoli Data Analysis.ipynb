# Purchasing Analysis (Gender)

```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

#Create a summary data frame to hold the results

#Optional: give the displayed data cleaner formatting

#Display the summary data frame
```

```python
df.head()
```

```python
#first I get the average price per gender
avprice=pd.pivot_table(df,index=["Gender"],values=["Price"])
avprice
```

```python
#Then the total purchase per Gender
tpg=pd.pivot_table(df,index=["Gender"],values=["Price"],aggfunc=[np.sum,len])
tpg
```

```python
#I get the avg purchase by person
merge_table=pd.merge(avprice, tpg, on="Gender")
merge_table
```

```python
merge_table.columns=['Average Purchase Price', 'Total Purchase Value', 'Purchase Count']
```

```python
merge_table['Avg Total Purchase per Person'] = merge_table['Total Purchase Value']/merge_table['Purchase Count']
```

#  Age Demographics

```python
#Establish bins for ages
#Categorize the existing players using the age bins. Hint: use pd.cut()
#Calculate the numbers and percentages by age group
#Create a summary data frame to hold the results
#Optional: round the percentage column to two decimal points
#Display Age Demographics Table
```
